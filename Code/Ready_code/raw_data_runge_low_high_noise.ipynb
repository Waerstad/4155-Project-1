```python
# Imports
import numpy as np
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error

from sklearn.preprocessing import StandardScaler
from sklearn.utils import resample
```

```python
# The function used to generate data, the Runge function
def Runge_func(x):
    return 1.0/(1 + 25*x**2)
```

```python
# Setting up dataset
np.random.seed(2025)
n = 1000
std_data_01 = 0.1
std_data_05 = 0.5
x = np.linspace(-1, 1, n)
y01 = Runge_func(x) + np.random.normal(0, std_data_01, size=n)
y05 = Runge_func(x) + np.random.normal(0, std_data_05, size=n)

# This configuration is tested and optimized for use in the latex report, even tho it looks a bit off here
plt.plot(x, y05, '.', markersize="1.5", label=f"Data (std. {std_data_05})")
plt.plot(x, y01, ".", markersize="1.5", label=f"Data (std. {std_data_01})")
plt.plot(x, Runge_func(x), "b-", label="Runge function", linewidth="0.75")
plt.title(f"Two typical iterations of the data \n" 
          f"with {n} points, and the Runge function \n"
           "used to generate it \n"
           "(with normally distributed noise added)"
          )
plt.xlabel("x")
plt.ylabel("Value")
plt.legend().get_frame().set_alpha(1.0) # Add solid background (non see thru) to legend box
plt.grid(True, which="both", linestyle="--", linewidth=0.8)
plt.show()
plt.close()
```

```python
data_to_save = np.c_[x, y01, y05]

np.savetxt("data_rawdata_lownoise_highnoise.out", data_to_save, header="x values, y std. 0.1, y std. 0.5 ;parameters: n=1000,")
```
